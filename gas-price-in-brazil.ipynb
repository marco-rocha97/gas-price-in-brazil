### 1. Uploading both data files and combining into one DataFrame.

```python
df1 = pd.read_csv('datasets/gasolina_2000+.csv', index_col=0, low_memory=False)
df2 = pd.read_csv('datasets/gasolina_2010+.csv', index_col=0, low_memory=False)
df = pd.concat([df1, df2])
df
```

### 4. Let's convert to Datetime format using pd.to_datetime().

```python
df['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'], format='%Y-%m-%d')
df['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'], format='%Y-%m-%d')
df.info()
```

### 5. Creating a new column to represent year and month as aaaa-mm, using DATA FINAL as reference.

```python
'''
To_period have 'M' function that creates the date format I want.
But for a specific case, it's also possible to use apply function.
'''
df['ANO MES'] = pd.to_datetime(df['DATA FINAL']).dt.to_period('M')
df
```

### 6. Listing all product types using value_counts().

```python
df['PRODUTO'].value_counts()
```

### 7. Creating a new DataFrame to analyze 'GASOLINA COMUM' data.

```python
df_gas_comum = df[df['PRODUTO'] == 'GASOLINA COMUM'].copy()
df_gas_comum
```

### 8. What was the average retail price of gasoline in August 2008?

```python
df_aug_08 = df_gas_comum[(df_gas_comum['ANO MES'] == '2008-08')]
df_aug_08['PREÇO MÉDIO REVENDA'].mean()
```

### 9. What was the average retail price of gasoline in São Paulo in May 2014?

```python
df_may_14_sp = df_gas_comum[(df_gas_comum['ANO MES'] == '2014-05') & (df_gas_comum['ESTADO'] == 'SAO PAULO')]
df_may_14_sp['PREÇO MÉDIO REVENDA'].mean()
```

### 10. In which state(s) did gasoline break the R$5.00 barrier? And when did this happen?

```python
df_up_5 = df_gas_comum[df_gas_comum['PREÇO MÍNIMO REVENDA'] > 5]
df_up_5[['DATA INICIAL', 'ESTADO', 'PREÇO MÍNIMO REVENDA']].sort_values('DATA INICIAL')
```

### 11. What was the average price in the southern states in 2012?

```python
df_2012_south = df_gas_comum[(df_gas_comum['DATA INICIAL'].dt.year == 2012) & (df_gas_comum['REGIÃO'] == 'SUL')]
df_2012_south['PREÇO MÉDIO REVENDA'].mean()
```

### 12. Obtain a table containing the year over year percentage change for the state of Rio de Janeiro

#### Filtering table and creating a month column

```python
df_rj = df_gas_comum[(df_gas_comum['ESTADO'] == 'RIO DE JANEIRO') & (df_gas_comum['PRODUTO'] == 'GASOLINA COMUM')].copy()
df_rj['MÊS'] = df_rj['DATA FINAL'].dt.month
df_rj
```

#### Grouping last month value (closing price) and filtering december to compare YoY

```python
df_rj = df_rj.groupby('ANO MES')[['PREÇO MÉDIO REVENDA', 'MÊS']].last()
df_rj = df_rj[df_rj['MÊS'] == 12]
df_rj
```

```python
df_rj['VAR'] = df_rj['PREÇO MÉDIO REVENDA']/df_rj['PREÇO MÉDIO REVENDA'].shift(1) - 1
df_rj
```

### CHALLENGE: Create a table containing a time series with the absolute and percentage difference between the cheapest and most expensive prices. Also show the states in which the highest and lowest prices were recorded.

#### Filtering the df and calcutating max and min values

```python
df_chlng = df[df['PRODUTO'] == 'GASOLINA COMUM']
df_max = df_chlng.groupby('ANO MES').max()['PREÇO MÉDIO REVENDA']
df_min = df_chlng.groupby('ANO MES').min()['PREÇO MÉDIO REVENDA']

df_chlng
```

```python
df_max
```

```python
df_min
```

#### Calculating absolute and percentage differences

```python
df_diff = pd.DataFrame()
df_diff['abs'] = df_max - df_min
df_diff['perc'] = df_diff['abs'] / df_min

df_diff
```

#### Relating max and min prices for each state with the row id

```python
idx_max = df_chlng.groupby('ANO MES')['PREÇO MÉDIO REVENDA'].idxmax()
idx_min = df_chlng.groupby('ANO MES')['PREÇO MÉDIO REVENDA'].idxmin()
```

```python
df_diff['max'] = idx_max
df_diff['min'] = idx_min

df_diff
```

```python
df_diff['estado_max'] = df_chlng.loc[idx_max, :]['ESTADO'].values
df_diff['estado_min'] = df_chlng.loc[idx_min, :]['ESTADO'].values

df_diff
```
